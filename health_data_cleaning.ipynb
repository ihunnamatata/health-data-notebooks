# 🧪 Health Data Cleaning Notebook
This notebook walks through cleaning and preparing electronic medical record (EMR) data using Python and Pandas.

```python
import pandas as pd

# Load sample data
df = pd.read_csv('https://raw.githubusercontent.com/datablist/sample-csv-files/main/files/people/people-100.csv')
df.head()
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Remove duplicates
df = df.drop_duplicates()

# Fill missing values
df.fillna('N/A', inplace=True)
df.head()
```
